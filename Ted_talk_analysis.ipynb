```python
import ast
```

```python

ted['ratings_list']=ted['ratings'].apply(lambda x: ast.literal_eval(x))
```

```python
ted.head()
```

```python
e=[]
for i in range(0,len(ted['ratings_list'])):
    c=0
    d=0
    for j in range(0,len(ted['ratings_list'][i])):
        c=ted['ratings_list'][i][j]['count']
        d=d+c
    e.append(d)
    
ted['num_ratings']=e
```

```python
ted['num_ratings'].describe()
```

```python
ted['ratings_list'][0]
```

```python
e=[]
for i in range(0,len(ted['ratings_list'])):
    c=0
    d=0
    for j in range(0,len(ted['ratings_list'][i])):
        if ted['ratings_list'][i][j]['name']=='Funny'
        d=d+c
    e.append(d)
    
ted['num_ratings']=e
```

```python
ted['ratings'].str.contains('Funny').value_counts()
```

```python
type(ted['ratings_list'][0])
```

```python
type(ted['ratings'][0])
```

```python
def get_funny(x):
    for d in x:
        if d['name']=='Funny':
            return d['count']
```

```python
ted['funny_ratings']=ted['ratings_list'].apply(get_funny)
```

```python
ted['funny_ratings'].head()
```

```python
ted['funny_rate']=ted['funny_ratings']/ted['num_ratings']
```

```python
ted.sort_values('funny_ratings')['speaker_occupation'].tail(20)
```

```python
ted.groupby('speaker_occupation')['funny_rate'].agg(['count','mean']).sort_values('mean').tail()
```
